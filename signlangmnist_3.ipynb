```python
import numpy as np 
import pandas as pd 
import time
from matplotlib import pyplot as plt 
from sklearn.metrics import confusion_matrix 
from  sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB 
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance



data_train = pd.read_csv('../input/sign-language-mnist/sign_mnist_train/sign_mnist_train.csv')
data_test = pd.read_csv('../input/sign-language-mnist/sign_mnist_test/sign_mnist_test.csv')



print('train data shape: ',data_train.shape)
print('test data shape: ',data_test.shape)
#data_train.head()
```

```python
data_train = np.array(data_train)
data_test = np.array(data_test)
m, n = data_train.shape
print('rows', m)
print('cols', n)
np.random.shuffle(data_train)
#data_dev = data_train[0:1000].T
#print('dataDev',data_dev.shape)

Y_test = data_test[:, 0]



#Training data
X_train = data_train[:, 1:]
X_train = X_train / 255
X_train=np.array(X_train)

Y_train = data_train[:, 0]
Y_train=np.array(Y_train)

X_test = data_test[:, 1:]
X_test = X_test / 255
X_test=np.array(X_test)

Y_test = data_test[:, 0]
Y_test=np.array(Y_test)





#print('X_train',X_train)
#print('Y_train',Y_test)
```

```python
#random forest change n_estimator to get better results
times = []
accuracies = []
start_time = time.time()
one_tree = RandomForestClassifier(n_jobs=-1, n_estimators=1)
one_tree.fit(X_train, Y_train)
one_score= one_tree.score(X_test, Y_test)
print('1 tree score',one_score)
elapsed_time = time.time() - start_time
print('elapsed time 1 tree',elapsed_time)
times.append(elapsed_time)
accuracies.append(one_score)

start_time = time.time()
ten_tree = RandomForestClassifier(n_jobs=-1, n_estimators=10)
ten_tree.fit(X_train, Y_train)
ten_score= ten_tree.score(X_test, Y_test)
print('10 trees score',ten_score)
elapsed_time = time.time() - start_time
print('elapsed time 10 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(ten_score)

start_time = time.time()
twenty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=20)
twenty_tree.fit(X_train, Y_train)
twenty_score= twenty_tree.score(X_test, Y_test)
print('20 trees score',twenty_score)
elapsed_time = time.time() - start_time
print('elapsed time 20 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(twenty_score)

start_time = time.time()
forty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=40)
forty_tree.fit(X_train, Y_train)
forty_score= forty_tree.score(X_test, Y_test)
print('40 trees score',forty_score)
elapsed_time = time.time() - start_time
print('elapsed time 40 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(forty_score)

start_time = time.time()
eighty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=80)
eighty_tree.fit(X_train, Y_train)
eighty_score= eighty_tree.score(X_test, Y_test)
print('80 trees score',eighty_score)
elapsed_time = time.time() - start_time
print('elapsed time 80 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(eighty_score)

start_time = time.time()
onesixty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=160)
onesixty_tree.fit(X_train, Y_train)
onesixty_score= onesixty_tree.score(X_test, Y_test)
print('160 trees score',onesixty_score)
elapsed_time = time.time() - start_time
print('elapsed time 160 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(onesixty_score)

start_time = time.time()
threetwenty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=320)
threetwenty_tree.fit(X_train, Y_train)
threetwenty_score= threetwenty_tree.score(X_test, Y_test)
print('320 trees score',threetwenty_score)
elapsed_time = time.time() - start_time
print('elapsed time 320 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(threetwenty_score)

start_time = time.time()
sixfourty_tree = RandomForestClassifier(n_jobs=-1, n_estimators=640)
sixfourty_tree.fit(X_train, Y_train)
sixfourty_score= sixfourty_tree.score(X_test, Y_test)
print('640 trees score',sixfourty_score)
elapsed_time = time.time() - start_time
print('elapsed time 640 trees',elapsed_time)
times.append(elapsed_time)
accuracies.append(sixfourty_score)

#accuracies
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
number_trees = ['1 tree ','10 trees' ,'20 trees','40 trees','80 trees','160 trees','320 trees','640 trees']
ax.bar(number_trees,accuracies)
plt.show()

#times
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
number_trees = ['1 tree ','10 trees' ,'20 trees','40 trees','80 trees','160 trees','320 trees','640 trees']
ax.bar(number_trees,times)
plt.show()

```
